```python
import numpy as np
import pandas as pd
# import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings; warnings.filterwarnings(action='once')
%matplotlib inline
```

```python
# Загружаем датасет
df = pd.read_csv("2019.csv")
```

```python
# Проверка на кореляции, каждого столбца, с каждым
sns.heatmap(df.corr(),annot=True,cmap='RdYlGn',linewidths=0.2) #df.corr()-->матрица корреляций
fig=plt.gcf()
fig.set_size_inches(12,12) # Размер матрицы в дюймах
plt.show()
```

Получаем оценку от -1 до 1 по сравнению всех полей со всеми.
Чем ближе к 1, тем больше корреляция между полями.
Например "ВВП на душу населения" и "Ожидаемая продолжительность здоровой жизни" имеют индекс 0.84
Есть нюансы! Общий рейтинг, чем больше цифра, тем рейтинг хуже. И это может запутать! Математически рейтинг выше и имеет обратную корреляцию с оценкой опрашиваемых -0.99

В общем используйте, но помните, что корреляции — обманная штука!
